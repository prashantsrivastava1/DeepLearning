```python
import tensorflow as tf
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("house_price_full.csv")
df.head()
```

```python
X = df.copy()
Y = X.pop('price')
X, Y
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)

Y = np.log(Y)

X,Y
```

```python
df_scaled = pd.DataFrame(X)
df_scaled
```

```python
x1,x2 = df_scaled.iloc[0]
x1,x2
```

```python
x = tf.constant([[x1,x2]],dtype=tf.float32)
x
```

```python
w1 = tf.Variable([0.2],dtype=tf.float32)
w2 = tf.Variable([0.15],dtype=tf.float32)
b = tf.Variable([0.1],dtype=tf.float32)
z = w1*x1 + w2*x2 + b
h = tf.math.sigmoid(z)
print("Output from 1st neuron is: ",h)
```

```python
## layer1 weights
# neuron1
b1 = tf.Variable([0.1])
w11 = tf.Variable([0.2])
w12 = tf.Variable([0.15])
#neuron2
b2 = tf.Variable([0.25])
w21 = tf.Variable([0.5])
w22 = tf.Variable([0.6])
## forward pass
# neuron 1
z1 = b1+w11*x1+w12*x2
h1 = tf.math.sigmoid(z1)
print("The output from the first neuron is",h1)
## forward pass
# neuron 2
z2 = b2+w21*x1+w22*x2
h2 = tf.math.sigmoid(z2)
print("The output from the second neuron is",h2)
## layer2 weights
b1 = tf.Variable([0.4])
w11 = tf.Variable([0.3])
w12 = tf.Variable([0.2])
## forward pass
# second layer
z1 = b1+w11*h1+w12*h2
h1 = z1
print("The output from the first neuron is",h1)
```

```python
y_true = Y[0]
y_pred = h1.numpy()
```

```python
#loss
L = 0.5*(y_true - y_pred)**2
print("The MSE error is",L)
```
